```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
votacao_parlamentar = pd.read_csv('../data/votacao_parlamentar.csv')
orgao_deputado = pd.read_csv('../data/orgao_deputado.csv')
proposicao_microdados = pd.read_csv('../data/proposicao_microdados.csv')
```

```python
# Average number of votes per proposition
votes_per_proposition = votacao_parlamentar.groupby('id_votacao')['voto'].count()
average_votes_per_proposition = votes_per_proposition.mean()

plt.figure(figsize=(10, 6))
sns.histplot(votes_per_proposition, bins=30, kde=True)
plt.axvline(average_votes_per_proposition, color='r', linestyle='--')
plt.title('Distribution of Votes per Proposition')
plt.xlabel('Number of Votes')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the number of votes for each proposition
votes_per_proposition = votacao_parlamentar.groupby('id_votacao').size()

# Filter propositions with more than 200 votes
propositions_with_more_than_200_votes = votes_per_proposition[votes_per_proposition > 200].index

# Filter the dataset to include only these propositions
filtered_votacao_parlamentar = votacao_parlamentar[votacao_parlamentar['id_votacao'].isin(propositions_with_more_than_200_votes)]

filtered_votacao_parlamentar
```

```python
# Convert the "data" column to datetime
votacao_parlamentar['data'] = pd.to_datetime(votacao_parlamentar['data'])

# Extract the year from the "data" column
votacao_parlamentar['year'] = votacao_parlamentar['data'].dt.year

# Drop duplicate propositions based on "id_votacao" and "year"
unique_propositions_per_year = votacao_parlamentar.drop_duplicates(subset=['id_votacao', 'year'])
```

```python
# Count the number of unique propositions per year
propositions_count_per_year = unique_propositions_per_year['year'].value_counts().sort_index()

# Plotting the count of unique propositions per year
plt.figure(figsize=(10, 6))
sns.barplot(x=propositions_count_per_year.index, y=propositions_count_per_year.values)
plt.title('Count of Unique Propositions per Year')
plt.xlabel('Year')
plt.ylabel('Count of Unique Propositions')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of votes for each deputy
votes_per_deputy = votacao_parlamentar['id_deputado'].value_counts()

# Plotting the histogram of votes by deputy ID
plt.figure(figsize=(15, 8))
sns.histplot(votes_per_deputy, bins=50, kde=True)
plt.title('Histogram of Votes by Deputy ID')
plt.xlabel('Number of Votes')
plt.ylabel('Frequency')
plt.show()
```
