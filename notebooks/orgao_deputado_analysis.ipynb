```python
# Filter for unique values of "nome" and "sigla"
unique_nome_sigla = orgao_deputado[['nome', 'sigla']].drop_duplicates()

unique_nome_sigla
```

```python
# List of relevant siglas based on manual inspection
relevant_siglas = [
    "PLEN",  # Plenário
    "MESA", 
    "CN",    # Congresso
]

# Filter the dataset to include only the relevant siglas
filtered_orgao_deputado = unique_nome_sigla[unique_nome_sigla['sigla'].isin(relevant_siglas)]
filtered_orgao_deputado
```
